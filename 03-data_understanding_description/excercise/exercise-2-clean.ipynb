# Ejemplo 2. Limpieza NaNs 



```python
import pandas as pd
```

El método `read_csv()` de **Pandas** se utiliza para leer archivos **CSV** (Comma-Separated Values) y convertirlos en un **DataFrame**. Es una de las funciones más utilizadas para la carga de datos en análisis y ciencia de datos.

```python
df = pd.read_csv('../../datasets/melbourne_housing-raw.csv', sep=',')
df
```

```python
df.shape
```

```python
df.head(5)
```

```python
df.tail(5)
```

El método `isna()` de **Pandas** se utiliza para identificar valores nulos o faltantes en un **DataFrame** o una **Serie**. Devuelve un objeto del mismo tamaño (un **DataFrame** o **Serie**) con valores booleanos: `True` si el elemento es nulo y `False` si no lo es.

El método `sum()` en **Pandas** se utiliza para calcular la suma de los elementos a lo largo de un **DataFrame** o una **Serie**. Es una herramienta poderosa y flexible para realizar operaciones de agregación, ya sea por filas o por columnas.

```python
df.isna().sum()
```

Existen demasiados `NaNs` en las columnas `'BuildingArea'` y `'YearBuilt'`, así que vamos a deshacernos de esas columnas.

El método `drop()` en Pandas se utiliza para eliminar filas o columnas de un **DataFrame**. 

```python
df_2 = df.drop(columns=['BuildingArea', 'YearBuilt'])
df_2.isna().sum()
```

Llenemos los `NaNs` en `Regionname` con la palabra `Unknown`:

```python
df_2['Regionname'] = df_2['Regionname'].fillna('Unknown')
df_2.isna().sum()
```

Ahora eliminemos las filas que contengan al menos un `NaN`:

```python
df_dropped = df_2.dropna(axis=0, how='any')
df_dropped.isna().sum()
```

Éste es el número de columnas con el que nos quedamos:

```python
df_dropped.shape
```

## Reindexando y renombrando columnas

Tenemos ahora un dataset que ha sido limpiado de `NaNs`. Tenemos ahora dos problemas. El primero es que nuestro índice no corresponde al número de filas que tenemos ahora:

<div style="padding: 10px; margin: 20px"><img src='./imgs/sesion-5_40.png'></div>



El método `reset_index()` en **Pandas** se utiliza para restablecer el índice de un **DataFrame** o una **Serie** a su configuración predeterminada. Este método es particularmente útil cuando deseas convertir un índice actual en una columna normal o para reorganizar el índice después de operaciones que modificaron filas o índices.

```python
df_dropped.reset_index()
```

Nuestro índice ya está correcto, pero ahora tenemos un columna llamada `index` que contiene el índice original. Como no queremos guardar esos datos, agregamos la opción `drop=True` para eliminar el índice anterior:

```python
df_dropped.reset_index(drop=True)
```

```python
df_dropped = df_dropped.reset_index(drop=True)
```

El segundo es que los nombres de nuestras columna son muy inconsistentes (e incluso contienen errores ortográficos):

<div style="padding: 10px; margin: 20px"><img src='./imgs/sesion-5_46.png'></div>

```python
column_name_mapping = {
    'Suburb': 'suburb',
    'Address': 'address',
    'Rooms': 'rooms',
    'Type': 'type',
    'Price': 'price',
    'Method': 'method',
    'SellerG': 'seller_g',
    'Date': 'date',
    'Distance': 'distance',
    'Postcode': 'post_code',
    'Bedroom2': 'bedrooms',
    'Bathroom': 'bathroom',
    'Car': 'car',
    'Landsize': 'land_size',
    'CouncilArea': 'council_area',
    'Lattitude': 'latitude',
    'Longtitude': 'longitude',
    'Regionname': 'region_name',
    'Propertycount': 'property_count'
}
```

```python
df_renamed = df_dropped.rename(columns=column_name_mapping)
df_renamed
```
