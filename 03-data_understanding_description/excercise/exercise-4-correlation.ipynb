# Relación entre las variables

```python
import pandas as pd
```

```python
df = pd.read_csv('../../datasets/melbourne_housing-clean.csv', index_col=0)
df.head()
```

El método `df.corr()` en Pandas calcula la correlación entre las columnas numéricas de un **DataFrame**. La correlación es una medida estadística que indica el grado de relación entre dos variables, y su valor está comprendido entre -1 y 1:
- 1: Correlación positiva perfecta (cuando una variable aumenta, la otra también lo hace de manera proporcional).
- 0: No hay correlación (las variables no tienen relación lineal).
- -1: Correlación negativa perfecta (cuando una variable aumenta, la otra disminuye proporcionalmente).

```python
# Correlación entre variables numéricas
numerical_df = df.select_dtypes(include=['float64', 'int64'])
print(numerical_df.corr())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

corr = numerical_df.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Matriz de Correlación')
plt.show()
```
